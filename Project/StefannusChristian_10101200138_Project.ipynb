### Mencari nilai rata-rata dan standar deviasi  dari wine points dan wine price
<br>
Untuk memudahkan pekerjaan, saya membuat fungsi rata-rata dan standar deviasi yang akan digunakan untuk mencari rata-rata dan standar deviasi dari wine points dan wine price

```python
wine_points = wine_review.points
wine_price = wine_review.price

def rata_rata(values):
  return sum(values)/len(values)
def standar_deviation(values):
    length = len(values)
    r = rata_rata(values)
    total_sum = 0
    for i in (values):    
        total_sum += (i-r)**2
    return math.sqrt(total_sum/length)
```

### Menyimpan dan memprint rata-rata points dan price
<br>
Disini untuk memudahkan pekerjaan, saya menyimpan rata-rata dari masing - masing wine price dan wine points kedalam variabel baru 
<br>
Kemudian saya memprint hasil dari rata-rata dan standar deviasi dari masing-masing wine points dan wine price serta menambah dan mengurang standar deviasi 

```python
rata_rata_wine_price = rata_rata(wine_price)
rata_rata_wine_points = rata_rata(wine_points)
standar_deviation_wine_points = standar_deviation(wine_points)
standar_deviation_wine_price = standar_deviation(wine_price)
rata_points_tambah_standev_points = rata_rata_wine_points + standar_deviation_wine_points
rata_points_kurang_standev_points = rata_rata_wine_points - standar_deviation_wine_points
rata_price_tambah_standev_price = rata_rata_wine_price + standar_deviation_wine_price
rata_price_kurang_standev_price = rata_rata_wine_price - standar_deviation_wine_price

print("Data wine points")
print("Rata-rata dari wine points adalah",rata_rata_wine_points)
print("Standar Deviasi dari wine points adalah",standar_deviation_wine_points)
print("Rata-rata wine points + standar deviasi points adalah ",rata_points_tambah_standev_points)
print("Rata-rata wine points - standar deviasi points adalah ",rata_points_kurang_standev_points)
print("\n")
print("Data wine price")
print("Rata-rata dari wine price adalah",rata_rata_wine_price)
print("Standar Deviasi dari wine price adalah",standar_deviation_wine_price)
print("Rata-rata wine price + standar deviasi price adalah ",rata_price_tambah_standev_price)
print("Rata-rata wine price - standar deviasi price adalah ",rata_price_kurang_standev_price)
```

### Membuat Histogram Wine Points dengan tambahan garis rata - rata dan rata-rata +- standar deviasi
<br>
Disini saya menampilkan histogram dengan sumbu x sebagai rata - rata wine points dan sumbu y sebagai jumlah wine

```python
sns.set(rc={'figure.figsize':(10,10)})
histogram_wine_points = sns.displot(wine_points)
plt.axvline(rata_rata_wine_points,color='m',label="rata-rata")
plt.axvline(rata_points_tambah_standev_points,color='g',label="rata-rata + standev")
plt.axvline(rata_points_kurang_standev_points,color='y',label="rata-rata - standev")
plt.legend()
plt.show()
```
